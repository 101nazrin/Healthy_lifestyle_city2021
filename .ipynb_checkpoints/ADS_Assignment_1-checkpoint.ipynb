### Function Defentions

```python
"""Take the Dataframe colums as series and fill the Nan values(-) with median of corresponding column"""
def Fill_value(series):
    return series.fillna(series.median())
```

```python
"""Takes the dataframe and numerical colum to calculate the corresponding data descriptions like mean,standard Deviation,
   minimum and maximum values within the data"""
def Describe(data_frame, numerical_column):
    min=(data_frame[numerical_column].min())
    max=(data_frame[numerical_column].max())
    mean=(data_frame[numerical_column].mean())
    std=(data_frame[numerical_column].std())
    df1 = pd.DataFrame(data={ 'mean': mean, 'std ': std,
                      'maximum in ': max,'minimum in' : min},index=['*'])
    print(df1)
    return df1
```

```python
"""used to plot the define the labels,fontsize and rotation of xticks in plotting """
def plot_labels(x_label,y_label,title):
    plt.xticks( rotation='vertical')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title,fontsize= 15, fontweight='bold')
```

## Data Cleaning

```python
#Removing Colums with less than 90% of data
thresho=len(data)*.1
thresho
data.dropna(thresh=thresho,axis=1,inplace=True)
print(data.isnull().sum())
```

```python
##Filling missing values of data with the median of the series
data.Annualavg_hours_worked = data['Annualavg_hours_worked'].transform(Fill_value)
data.Sunshine_hours_City = data['Sunshine_hours_City'].transform(Fill_value)
data.Pollution_Index_score_City = data['Pollution_Index_score_City'].transform(Fill_value)
```

```python
data['Pollution_Index_score_City'].median()
```

```python
#Removing the Character 'Â£' from the numeric columns of given data
data['Cost _of_bottle_of_water(City)']=data['Cost _of_bottle_of_water(City)'].apply(lambda x: str(x).replace('Â£','')
                                                                             if 'Â£' in str(x) else str(x))
data['Cost _of_bottle_of_water(City)']=data['Cost _of_bottle_of_water(City)'].apply(lambda x: float(x))


data['Cost_of_a_monthly_gym_membership(City)']=data['Cost_of_a_monthly_gym_membership(City)'].apply(lambda x: str(x).
                                                                             replace('Â£','')if 'Â£' in str(x) else str(x))
data['Cost_of_a_monthly_gym_membership(City)']=data['Cost_of_a_monthly_gym_membership(City)'].apply(lambda x: float(x))
```
